```python
import pandas as pd
import matplotlib.pyplot as plt
% matplotlib inline
recent_grads = pd.read_csv('recent-grads.csv')
print(recent_grads.iloc[0])
print(recent_grads.head())
print(recent_grads.tail())

#⚠️ the attribute of head(),tail()
```

### Look Up the number of rows/columns in Dataframe

```python
# ⚠️算dataframe的行数，取一列求长度即可
len(recent_grads['Rank'])
# ⚠️which equals to 
len(recent_grads)
# ⚠️ which equals to
recent_grads.shape

# ⚠️ for in loop 对dataframe来说是column loop 
count = 0
for column in recent_grads:
    count +=1
count
```

### Clean missing value

```python
recent_grads = recent_grads.dropna()
len(recent_grads['Rank'])
```

### Summary Statistics

```python
recent_grads.describe()
```

### Pandas, Histograms 2 methods

```python
recent_grads['Sample_size'].plot(kind='hist')
# ⚠️which does not equal to 
# recent_grads.plot('Sample_size',kind='hist')这种方法会把所有的column都用hist表示出来
```

```python
#⚠️ which equals to 
recent_grads['Sample_size'].hist(bins=25, range=(0,5000))
#the method allows us to customize such as bins, more feasible 
```

### Pandas, Scatter Plot

```python
recent_grads.plot(x='Sample_size', y='Median', kind='scatter')
# ⚠️无 recent_grads.scatter(x='Sample_size', y='Median')
# ⚠️只有一个variable放df后面，两个variable(即x&y)放括号里面
```

### Pandas, Scatter Matrix Plot

```python
from pandas.tools.plotting import scatter_matrix
scatter_matrix(recent_grads[['Sample_size', 'Median']], figsize=(6,6))
```

```python
scatter_matrix(recent_grads[['Sample_size', 'Median', 'Unemployment_rate']], figsize=(10,10))
```
