```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
reviews = pd.read_csv('fandango_scores.csv')
norm_reviews = reviews[['FILM','RT_user_norm','Metacritic_user_nom','IMDB_norm','Fandango_Ratingvalue','Fandango_Stars']]
norm_reviews.head(1)
```

### Create Bar Plots with plt.bar (left, height, width)

```python
# ⚠️shortcut of plt.figure() + fig.add_subplot()
fig, ax = plt.subplots()
num_cols = ['RT_user_norm', 'Metacritic_user_nom', 'IMDB_norm', 'Fandango_Ratingvalue', 'Fandango_Stars']
bar_heights = norm_reviews[num_cols].iloc[0].values
# ⚠️ don't forget to add 'values'
bar_positions = np.arange(5) + 0.75
ax.bar(bar_positions, bar_heights, width=0.5)
plt.show()
```

### customizing bar plots

```python
fig, ax = plt.subplots()
num_cols = ['RT_user_norm', 'Metacritic_user_nom', 'IMDB_norm', 'Fandango_Ratingvalue', 'Fandango_Stars']
bar_heights = norm_reviews[num_cols].iloc[0].values
bar_positions = np.arange(5) + 0.75
tick_positions = np.arange(1,6)
ax.bar(bar_positions, bar_heights, width=0.5)
ax.set_xticks(tick_positions)
ax.set_xticklabels(num_cols,rotation =90)
ax.set_xlabel('Rating Source')
ax.set_ylabel('Average Rating')
ax.set_title('Average User Rating For Avengers: Age of Ultron (2015)')
plt.show()

#⚠️ remember the methods 'set_xticks(tick_positions)'
#⚠️ remember the methods 'set_xticklabels()'
#⚠️ remember the methods 'set_xlabel()'
#⚠️ remember the methods 'set_title()'
```

### Horizontal Bar Plot with plt.barh( )

```python
# 只需要改成ax.barh(bottom=, width=,)
num_cols = ['RT_user_norm', 'Metacritic_user_nom', 'IMDB_norm', 'Fandango_Ratingvalue', 'Fandango_Stars']

bar_widths = norm_reviews[num_cols].iloc[0].values
bar_positions = np.arange(5) + 0.75
tick_positions = np.arange(1,6)

fig, ax = plt.subplots()
ax.barh(bar_positions, bar_widths, height=0.5)
ax.set_yticks(tick_positions)
ax.set_yticklabels(num_cols)
ax.set_ylabel("Rating Source")
ax.set_xlabel("Average Rating")
ax.set_title( "Average User Rating For Avengers: Age of Ultron (2015)")
plt.show()
```

### Benchmarking Correlation

```python
fig = plt.figure(figsize=(5,10))
ax1 = fig.add_subplot(3,1,1)
ax2 = fig.add_subplot(3,1,2)
ax3 = fig.add_subplot(3,1,3)

ax1.scatter(norm_reviews['Fandango_Ratingvalue'], norm_reviews['RT_user_norm'])
ax1.set_xlabel("Fandango")
ax1.set_ylabel("Rotten Tomatoes")
ax1.set_xlim(4,5)
ax1.set_ylim(4,5)

ax2.scatter(norm_reviews['Fandango_Ratingvalue'], norm_reviews['Metacritic_user_nom'])
ax2.set_xlabel("Fandango")
ax2.set_ylabel("Metacritic")
ax2.set_xlim(4,5)
ax2.set_ylim(4,5)


ax3.scatter(norm_reviews['Fandango_Ratingvalue'], norm_reviews['IMDB_norm'])
ax3.set_xlabel("Fandango")
ax3.set_ylabel("IMDB")
ax3.set_xlim(4,5)
ax3.set_ylim(4,5)

plt.show()
```

### Frequency Tabel

```python
# first value_counts, next sort_index()
frequency = norm_reviews['Fandango_Ratingvalue'].value_counts()
fandango_distribution = frequency.sort_index()

frequency2 = norm_reviews['IMDB_norm'].value_counts()
imdb_distribution = frequency2.sort_index()

print(fandango_distribution)
print(imdb_distribution)
```

### Histogram with plt.hist( variable, range, bins)

```python
fig = plt.figure(figsize=(5,20))
ax1 = fig.add_subplot(4,1,1)
ax2 = fig.add_subplot(4,1,2)
ax3 = fig.add_subplot(4,1,3)
ax4 = fig.add_subplot(4,1,4)

ax1.hist(norm_reviews['Fandango_Ratingvalue'], range=(0,5),bins= 20)
ax1.set_title('Distribution of Fandango Ratings')

ax2.hist(norm_reviews['RT_user_norm'], range=(0,5),bins= 20)
ax2.set_title('Distribution of Rotten Tomatoes Ratings')

ax3.hist(norm_reviews['Metacritic_user_nom'], range=(0,5),bins= 20)
ax3.set_title('Distribution of Metacritic Ratings')

ax4.hist(norm_reviews['IMDB_norm'], range=(0,5),bins= 20)
ax4.set_title('Distribution of IMDB Ratings')

ax1.set_ylim(0,50)
ax1.set_ylabel('Frequency')
ax2.set_ylim(0,50)
ax2.set_ylabel('Frequency')
ax3.set_ylim(0,50)
ax3.set_ylabel('Frequency')
ax4.set_ylim(0,50)
ax4.set_ylabel('Frequency')
plt.show()
```
