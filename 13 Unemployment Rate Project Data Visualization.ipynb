```python
import pandas as pd
unrate = pd.read_csv('unrate.csv')
unrate['DATE'] = pd.to_datetime(unrate['DATE'])
# ⚠️ rememer the method 'pd.to_datetime'
unrate.head(3)
```

### Line Charts

```python
import matplotlib.pyplot as plt
# line chart that visualizes the unemployment rates from 1948
first= unrate[0:12]
x_values = first['DATE']
y_values = first['VALUE']
plt.plot(x_values, y_values)
plt.xticks(rotation=90)
plt.xlabel('Month')
plt.ylabel('Unemployment Rate')
plt.title('Monthly Unemployment Trends, 1948')
plt.show()

#⚠️ remember the method 'xticks' & 'plot(x,y)' & 'xlabel' & 'ylabel'
```

### Multiple Plots

```python
fig = plt.figure()
#⚠️ axes_obj = fig.add_subplot(nrows, ncols, plot_number)
ax1 = fig.add_subplot(2,1,1)
ax2 = fig.add_subplot(2,1,2)
plt.show()
```

### Comparing Across More Years

```python
# In the top subplot, plot the data from 1948
# In the bottom subplot, plot the data from 1949
fig = plt.figure(figsize = (12,6))
# ⚠️ figsize parameter: specify the figure size
# ⚠️ ax.set_title (more 'set_'): subplot title function
ax1 = fig.add_subplot(2,1,1)
ax2 = fig.add_subplot(2,1,2)
ax1.plot(unrate[0:12]['DATE'], unrate[0:12]['VALUE'])
ax1.set_title('Monthly Unemployment Rate, 1948')
ax2.plot(unrate[12:24]['DATE'], unrate[12:24]['VALUE'])
ax2.set_title('Monthly Unemployment Rate, 1949')
plt.show()
```

### Overlaying Line Charts

```python
#⚠️ use this method to compare across months (把1949年和1950年的1-12月都弄成相同的x-axis)
unrate['MONTH'] = unrate['DATE'].dt.month
#⚠️ remember the parameter 'c', which means the color of lines
fig = plt.figure(figsize=(6,3))
plt.plot(unrate[0:12]['MONTH'], unrate[0:12]['VALUE'],c='red')
plt.plot(unrate[12:24]['MONTH'], unrate[12:24]['VALUE'],c='blue')
plt.show()
```

### Adding More Lines with for-in loop and A Legend

```python
fig = plt.figure(figsize=(10,6))
colors = ['red', 'blue', 'green', 'orange', 'black']
years = ['1948','1949','1950','1951','1952']
for i in range(5):
    start_index = i*12
    end_index = (i+1)*12
    subset = unrate[start_index:end_index]
    plt.plot(subset['MONTH'], subset['VALUE'], c=colors[i], label=years[i])
#⚠️plt.legend(loc) to create a legend for the whole plot
#⚠️use Axes.legend() to create a legend for each subplot
#⚠️legend需要和label配套使用，‘label' parameter is located in plt.plot()
plt.legend(loc='upper left')
plt.title("Monthly Unemployment Trends, 1948-1952")
plt.xlabel("Month, Integer")
plt.ylabel("Unemployment Rate, Percent")
plt.show()
```
